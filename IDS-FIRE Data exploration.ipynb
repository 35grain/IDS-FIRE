```python
with open('riiklikud_paastekomandod.json','r', encoding="UTF-8") as f:
    national_brigades_raw = json.loads(f.read())
    
with open('vabatahtlikud_paastekomandod.json','r', encoding="UTF-8") as f:
    voluntary_brigades_raw = json.loads(f.read())

national_brigades = pd.json_normalize(national_brigades_raw)
voluntary_brigades = pd.json_normalize(voluntary_brigades_raw)

# Remove unrelated data
national_brigades.drop(list(national_brigades.filter(regex = 'geometry.')), axis = 1, inplace = True)
voluntary_brigades.drop(list(voluntary_brigades.filter(regex = 'geometry.')), axis = 1, inplace = True)

national_brigades.columns = national_brigades.columns.str.replace('attributes.', '')
voluntary_brigades.columns = voluntary_brigades.columns.str.replace('attributes.', '')

national_brigades = national_brigades.drop('__OBJECTID', axis=1)
voluntary_brigades = voluntary_brigades.drop('__OBJECTID', axis=1)
```
