```python
import datetime
data = pd.concat([prev_years, current_year], ignore_index = True)

#String to DateTime
data['sundmuse_kuupaev_dt'] = pd.to_datetime(data['sundmuse_kuupaev_dt'])
data['hairekeskuses_registreerimine'] = pd.to_datetime(data['hairekeskuses_registreerimine'])
data['sundmuse_kuupaev_dt'] = pd.to_datetime(data['sundmuse_kuupaev_dt'])
data['hairekeskuses_registreerimine'] = pd.to_datetime(data['hairekeskuses_registreerimine'])
data['esimene_alarmeerimine'] = pd.to_datetime(data['esimene_alarmeerimine'])
data['esimene_valjasoit'] = pd.to_datetime(data['esimene_valjasoit'])
data['esimese_kohalejoudnud_aeg'] = pd.to_datetime(data['esimese_kohalejoudnud_aeg'])
data['lokaliseerimise_aeg'] = pd.to_datetime(data['lokaliseerimise_aeg'])
data['likvideerimise_aeg'] = pd.to_datetime(data['likvideerimise_aeg'])
data['viimase_ressursi_lahkumise_aeg'] = pd.to_datetime(data['viimase_ressursi_lahkumise_aeg'])
data['sundmuse_kestvus'] = data['sundmuse_kestvus'].astype(str)


#String to Numeric
data['wgs_latitude'] = pd.to_numeric(data['wgs_latitude'])
data['wgs_longitude'] = pd.to_numeric(data['wgs_longitude'])
data['valjasoidu_korralduste_arv'] = pd.to_numeric(data['valjasoidu_korralduste_arv'])
data['valjasoitnud_ressursside_arv'] = pd.to_numeric(data['valjasoitnud_ressursside_arv'])
data['annulleeritud_ressursside_arv'] = pd.to_numeric(data['annulleeritud_ressursside_arv'])
data['kohal_ressursside_arv'] = pd.to_numeric(data['kohal_ressursside_arv'])

data.dtypes
```

```python
def dt64_to_float(dt64):
     year = dt64.astype('M8[Y]')
     days = (dt64 - year).astype('timedelta64[D]')
     year_next = year + np.timedelta64(1, 'Y')
     days_of_year = (year_next.astype('M8[D]') - year.astype('M8[D]')).astype('timedelta64[D]')
     dt_float = 1970 + year.astype(float) + days / (days_of_year)
     return dt_float
df1 = data.copy()
df1['hairekeskuses_registreerimine'] = dt64_to_float(df1['hairekeskuses_registreerimine'].to_numpy())
df1['esimene_alarmeerimine'] = dt64_to_float(df1['esimene_alarmeerimine'].to_numpy())
df1['esimene_valjasoit'] = dt64_to_float(df1['esimene_valjasoit'].to_numpy())
df1['esimese_kohalejoudnud_aeg'] = dt64_to_float(df1['esimese_kohalejoudnud_aeg'].to_numpy())
df1['lokaliseerimise_aeg'] = dt64_to_float(df1['lokaliseerimise_aeg'].to_numpy())
df1['likvideerimise_aeg'] = dt64_to_float(df1['likvideerimise_aeg'].to_numpy())
df1['viimase_ressursi_lahkumise_aeg'] = dt64_to_float(df1['viimase_ressursi_lahkumise_aeg'].to_numpy())
df1['sundmuse_kuupaev_dt'] = dt64_to_float(df1['sundmuse_kuupaev_dt'].to_numpy())
```

```python
#df = data.copy()
df = df1.copy()
df.drop(df.loc[df['sundmuse_kestvus']=='nan'].index, inplace=True)
df = df.drop('viimase_ressursi_lahkumise_aeg', 1)
df = df.drop('lokaliseerimise_aeg', 1)
df = df.drop('likvideerimise_aeg', 1)

kestvused = df['sundmuse_kestvus'].to_numpy()

for x in range(len(kestvused)):
    kestvus = kestvused[x].split(":")
    tunnid = int(kestvus[0])
    minutid = int(kestvus[1])
    sekundid = int(kestvus[2])
    
    kokku = tunnid + minutid/60 + sekundid/3600
    
    kestvused[x] = kokku
    
df['sundmuse_kestvus'] = kestvused.tolist()
#df[df.select_dtypes(np.float64).columns] = df.select_dtypes(np.float64).astype(np.float32)
df.info()
```

```python
data_dum = pd.get_dummies(df, columns=['tulekahju_liik', 'sundmuse_alamliik', 'mis_poles', 'maakond', 'kov', 'korgem_aste'])
data_dum.fillna(data_dum.mean(), inplace=True)
```

```python
y = data_dum['sundmuse_kestvus']
X = data_dum.loc[:, data_dum.columns != 'sundmuse_kestvus']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score
import sklearn.metrics as sm

# Create linear regression object
regr = linear_model.LinearRegression()


# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)

print("Explain variance score =", round(sm.explained_variance_score(y_test, y_pred), 2))

# The coefficients
#print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred)) 
```

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=5)
neigh.fit(X_train, y_train)
y_preds = neigh.predict(X_test)

print("Coefficient of determination: %.2f" % r2_score(y_test, y_preds)) 
```

```python
from sklearn.linear_model import Ridge
#Fit the model
ridgereg = Ridge(alpha=1.0,normalize=True)
ridgereg.fit(X_train, y_train)
y_pred = ridgereg.predict(X_test)
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_pred), 2))
```

```python
from sklearn.linear_model import Lasso
#Fit the model
lassoreg = Lasso(alpha=1.0)
lassoreg.fit(X_train, y_train)
y_pred = lassoreg.predict(X_test)
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_pred), 2))
```

```python
from sklearn.linear_model import ElasticNet
regr = ElasticNet(random_state=0)
regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_pred), 2))
```
